```python
# ==========================================
# BAGIAN 1: CLOSE ADDRESSING (CHAINING)
# Hash Function: Modulo Biasa
# ==========================================
print("="*50)
print("1. CLOSE ADDRESSING (CHAINING) - CRUD")
print("="*50)

class ChainingHash:
    def __init__(self, size=5):
        self.size = size
        self.table = [[] for _ in range(self.size)]

    def hash_func(self, key):
        return key % self.size

    # CREATE & UPDATE (Jika key sudah ada, ganti value baru)
    def insert(self, key, value):
        idx = self.hash_func(key)
        for i, kv in enumerate(self.table[idx]):
            if kv[0] == key:
                self.table[idx][i] = (key, value) # Update
                print(f"[Update] Key {key} diperbarui menjadi {value}")
                return
        self.table[idx].append((key, value))      # Create
        print(f"[Insert] Key {key} masuk ke bucket {idx}")

    # READ
    def search(self, key):
        idx = self.hash_func(key)
        for kv in self.table[idx]:
            if kv[0] == key:
                return kv[1]
        return None

    # DELETE
    def delete(self, key):
        idx = self.hash_func(key)
        for i, kv in enumerate(self.table[idx]):
            if kv[0] == key:
                del self.table[idx][i]
                print(f"[Delete] Key {key} berhasil dihapus")
                return True
        print(f"[Delete] Key {key} tidak ditemukan")
        return False

    def display(self):
        print("Kondisi Tabel Chaining:")
        for i, bucket in enumerate(self.table):
            print(f"Bucket {i}: {bucket}")

# Test Chaining
chain_ht = ChainingHash(size=5)
chain_ht.insert(10, "A")
chain_ht.insert(22, "B")
chain_ht.insert(15, "C") # Tabrakan dengan 10 (15%5=0, 10%5=0)
chain_ht.insert(22, "Z") # Update data
print(f"Hasil Cari Key 15: {chain_ht.search(15)}")
chain_ht.delete(10)
chain_ht.display()


# ==========================================
# BAGIAN 2: OPEN ADDRESSING (LINEAR PROBING)
# Hash Function: Modulo Biasa
# ==========================================
print("\n" + "="*50)
print("2. OPEN ADDRESSING (LINEAR PROBING)")
print("Load Factor < 1 & Contoh Tabel Penuh")
print("="*50)

class LinearProbingHash:
    def __init__(self, size=5):
        self.size = size
        self.table = [None] * size
        self.count = 0

    def hash_func(self, key):
        return key % self.size

    def insert(self, key, value):
        if self.count >= self.size:
            print(f"[WARNING] Tabel Penuh! Load Factor >= 1. Tidak bisa insert ({key}: {value})")
            return

        idx = self.hash_func(key)
        awal_idx = idx
        while self.table[idx] is not None and self.table[idx] != "DELETED":
            if self.table[idx][0] == key: # Update
                self.table[idx] = (key, value)
                print(f"[Update] Key {key} diperbarui di index {idx}")
                return
            idx = (idx + 1) % self.size
            if idx == awal_idx:
                break
        
        self.table[idx] = (key, value)
        self.count += 1
        print(f"[Insert] Key {key} berhasil masuk ke index {idx}")

    def display(self):
        print("Tabel Linear Probing:", self.table)

# Test Linear Probing (Tabel ukuran 5, kita paksa isi 6 untuk lihat error load factor)
lp_ht = LinearProbingHash(size=5)
lp_ht.insert(1, "Satu")
lp_ht.insert(6, "Enam")   # Collision dengan 1, akan geser ke index sebelahnya
lp_ht.insert(11, "Sebelas")
lp_ht.insert(2, "Dua")
lp_ht.insert(3, "Tiga")
lp_ht.insert(99, "Penuh") # Ini akan memicu error tabel penuh!
lp_ht.display()


# ==========================================
# BAGIAN 3: OPEN ADDRESSING (QUADRATIC PROBING)
# Hash Function: MID SQUARE METHOD
# ==========================================
print("\n" + "="*50)
print("3. OPEN ADDRESSING (QUADRATIC PROBING)")
print("Hash: MID SQUARE METHOD")
print("="*50)

class QuadraticProbingHash:
    def __init__(self, size=7):
        self.size = size
        self.table = [None] * size
        self.count = 0

    # FUNGSI HASH 2: MID SQUARE METHOD
    def hash_mid_square(self, key):
        kuadrat = key * key
        str_k = str(kuadrat)
        mid = len(str_k) // 2
        # Ambil 1 digit di tengah
        mid_val = int(str_k[mid]) if len(str_k) > 0 else 0
        return mid_val % self.size

    def insert(self, key, value):
        if self.count >= self.size:
            print("Tabel Penuh!")
            return

        idx = self.hash_mid_square(key)
        i = 0
        # Quadratic Probing: (hash + i^2) % size
        new_idx = idx
        while self.table[new_idx] is not None and self.table[new_idx] != "DELETED":
            if self.table[new_idx][0] == key:
                self.table[new_idx] = (key, value)
                return
            i += 1
            new_idx = (idx + (i * i)) % self.size
            
        self.table[new_idx] = (key, value)
        self.count += 1
        print(f"[Insert] Key {key} (Kuadrat: {key**2}) -> Index Hash: {idx} -> Final Index: {new_idx}")

    def display(self):
        print("Tabel Quadratic Probing:", self.table)

# Test Quadratic Probing & Mid Square
qp_ht = QuadraticProbingHash(size=7)
qp_ht.insert(12, "Data A") # 12^2 = 144, tengah 4, 4%7 = 4
qp_ht.insert(25, "Data B") # 25^2 = 625, tengah 2, 2%7 = 2
qp_ht.insert(38, "Data C") # Uji bentrokan / quadratic lompat
qp_ht.insert(12, "Data A Baru") # Uji update
qp_ht.display()
print("="*50)
```
