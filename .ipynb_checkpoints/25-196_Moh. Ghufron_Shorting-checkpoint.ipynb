# Sorting - Perbandingan Waktu Eksekusi
## Bubble Sort | Selection Sort | Insertion Sort | Merge Sort | Quick Sort
### 100 Data Random, Urutan Ascending

```python
import time
import random

random.seed(42)
DATA_ASLI = random.sample(range(1, 1000), 100)

print("Data asli (100 elemen):")
print(DATA_ASLI)
print("Total:", len(DATA_ASLI))
```

```python
# ========================
# BUBBLE SORT
# ========================
def bubble_sort(arr):
    a = arr[:]
    n = len(a)
    for i in range(n):
        for j in range(0, n-i-1):
            if a[j] > a[j+1]:
                a[j], a[j+1] = a[j+1], a[j]
    return a

# ========================
# SELECTION SORT
# ========================
def selection_sort(arr):
    a = arr[:]
    n = len(a)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if a[j] < a[min_idx]:
                min_idx = j
        a[i], a[min_idx] = a[min_idx], a[i]
    return a

# ========================
# INSERTION SORT
# ========================
def insertion_sort(arr):
    a = arr[:]
    for i in range(1, len(a)):
        key = a[i]
        j = i - 1
        while j >= 0 and a[j] > key:
            a[j+1] = a[j]
            j -= 1
        a[j+1] = key
    return a

# --- Jalankan & Ukur Waktu ---
print("=" * 55)

# Bubble Sort
start = time.perf_counter()
hasil_bubble = bubble_sort(DATA_ASLI)
end = time.perf_counter()
waktu_bubble = end - start
print(f"Bubble Sort    : {waktu_bubble:.8f} detik")
print(f"Hasil          : {hasil_bubble}\n")

# Selection Sort
start = time.perf_counter()
hasil_selection = selection_sort(DATA_ASLI)
end = time.perf_counter()
waktu_selection = end - start
print(f"Selection Sort : {waktu_selection:.8f} detik")
print(f"Hasil          : {hasil_selection}\n")

# Insertion Sort
start = time.perf_counter()
hasil_insertion = insertion_sort(DATA_ASLI)
end = time.perf_counter()
waktu_insertion = end - start
print(f"Insertion Sort : {waktu_insertion:.8f} detik")
print(f"Hasil          : {hasil_insertion}\n")

print("=" * 55)
```

```python
# ========================
# MERGE SORT + DETAIL
# ========================
step = 0

def merge_sort(arr, level=0):
    global step
    indent = "  " * level

    print(f"{indent} SPLIT  : {arr}")

    if len(arr) <= 1:
        return arr

    mid   = len(arr) // 2
    left  = merge_sort(arr[:mid], level+1)
    right = merge_sort(arr[mid:], level+1)

    merged = merge(left, right)
    step += 1
    print(f"{indent} MERGE  : {left} + {right} → {merged}")
    return merged

def merge(left, right):
    result = []
    i = j  = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i]); i += 1
        else:
            result.append(right[j]); j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result

# --- Jalankan ---
print("=" * 55)
print("MERGE SORT — Detail Proses")
print("=" * 55)

# Jalankan Merge Sort (10 data biar detail keliatan)
sample_merge = DATA_ASLI[:10]

step = 0
start = time.perf_counter()
hasil_merge = merge_sort(sample_merge)
end = time.perf_counter()

waktu_merge = end - start

# Ukur waktu merge untuk 100 data
start = time.perf_counter()
merge_sort_100 = merge_sort(DATA_ASLI)
end = time.perf_counter()

waktu_merge_100 = end - start

print(f"Waktu Merge Sort (100 data): {waktu_merge_100:.8f} detik")
print("=" * 55)

print("\nHasil Merge Sort:", hasil_merge)
print("Total step:", step)
print(f"Waktu Merge (10 data): {waktu_merge:.8f} detik")
print("=" * 55)
# ========================
# QUICK SORT FUNCTION
# ========================
def quick_sort(arr, low, high):
    if low < high:
        pi = partition(arr, low, high)
        quick_sort(arr, low, pi-1)
        quick_sort(arr, pi+1, high)

def partition(arr, low, high):
    pivot = arr[high]
    i = low - 1
    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i+1], arr[high] = arr[high], arr[i+1]
    return i + 1

# ========================
# QUICK SORT 100 DATA
# ========================
data_quick_100 = DATA_ASLI[:]

start = time.perf_counter()
quick_sort(data_quick_100, 0, len(data_quick_100)-1)
end = time.perf_counter()

waktu_quick_100 = end - start
print(f"Waktu Quick Sort (100 data): {waktu_quick_100:.8f} detik")
print("Hasil Quick Sort (10 pertama):", data_quick_100[:10])
print("=" * 55)
```

```python
# ========================
# REKAP WAKTU SEMUA
# ========================
print("\n" + "=" * 50)
print("   REKAP WAKTU EKSEKUSI — 100 DATA")
print("=" * 50)
print(f"  Bubble Sort    : {waktu_bubble:.8f} detik")
print(f"  Selection Sort : {waktu_selection:.8f} detik")
print(f"  Insertion Sort : {waktu_insertion:.8f} detik")
print(f"  Merge Sort     : {waktu_merge_100:.8f} detik")
print(f"  Quick Sort     : {waktu_quick_100:.8f} detik")
print("=" * 50)

# Cari tercepat
hasil_waktu = {
    "Bubble Sort"   : waktu_bubble,
    "Selection Sort": waktu_selection,
    "Insertion Sort": waktu_insertion,
    "Merge Sort"    : waktu_merge_100,
    "Quick Sort"    : waktu_quick_100,
}
tercepat = min(hasil_waktu, key=hasil_waktu.get)
terlambat = max(hasil_waktu, key=hasil_waktu.get)
print(f"\n   Tercepat  : {tercepat}")
print(f"  Terlambat : {terlambat}")
```
